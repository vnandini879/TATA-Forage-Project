```python
# Drop rows where any column has a null value
df_cleaned = df.dropna().copy()

print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
```

```python
print("Info for df_cleaned:")
df_cleaned.info()
```

```python
df_cleaned['InvoiceDate'] = pd.to_datetime(df_cleaned['InvoiceDate'], errors='coerce')

# Check if any dates were coerced to NaT
nat_count = df_cleaned['InvoiceDate'].isnull().sum()
if nat_count > 0:
    print(f"Warning: {nat_count} invalid dates were converted to NaT.")
    # Optionally, you can drop these rows or fill them with a placeholder
    # df_cleaned.dropna(subset=['InvoiceDate'], inplace=True)

print("Updated info for df_cleaned after converting InvoiceDate:")
df_cleaned.info()

print("\nFirst 5 rows with updated InvoiceDate:")
display(df_cleaned)
```
